```python
# Conectar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Rutas de archivos
ruta_csv = "/content/drive/MyDrive/Intro Datos/datos.csv"
ruta_txt = "/content/drive/MyDrive/Intro Datos/texto.txt"
ruta_pkl = "/content/drive/MyDrive/Intro Datos/objeto.pkl"
```

## Ejercicio 1

Lectura y escritura de archivos CSV y TXT usando context managers


```python
import csv

# Leer TXT
with open(ruta_txt, "r", encoding="utf-8") as archivo:
    print("Contenido TXT:\n")
    print(archivo.read())

# Leer CSV
print("\nContenido CSV:\n")
with open(ruta_csv, "r", encoding="utf-8") as archivo:
    lector = csv.reader(archivo)
    for fila in lector:
        print(fila)

# Escribir TXT
with open("/content/drive/MyDrive/Intro Datos/salida.txt", "w", encoding="utf-8") as archivo:
    archivo.write("Archivo creado desde Colab\n")

# Escribir CSV
datos = [
    ["producto", "precio"],
    ["Laptop", 5000],
    ["Mouse", 150]
]

with open("/content/drive/MyDrive/Intro Datos/salida.csv", "w", encoding="utf-8", newline="") as archivo:
    escritor = csv.writer(archivo)
    escritor.writerows(datos)
```

### Pregunta

¿En qué parte o partes de un Pipeline de Datos se pueden usar las operaciones de Lectura (R) y Escritura (W)?

**Respuesta:**
Las operaciones de lectura (R) se utilizan al inicio del pipeline para obtener datos desde fuentes externas y también entre etapas.
Las operaciones de escritura (W) se utilizan al final del pipeline o para guardar resultados intermedios procesados.


## Ejercicio 2

Uso de Pandas para leer y escribir archivos CSV


```python
import pandas as pd

# Leer CSV en DataFrame
df = pd.read_csv(ruta_csv)
print("DataFrame:\n")
print(df)

# Escribir CSV desde DataFrame
df.to_csv("/content/drive/MyDrive/Intro Datos/salida_pandas.csv", index=False)
```

**Respuesta:**

Sin embargo, no es adecuado para datos no estructurados como imágenes, audio o video.


## Ejercicio 3

Lectura y escritura de archivos Pickle


```python
import pickle

# Leer Pickle
with open(ruta_pkl, "rb") as archivo:
    objeto = pickle.load(archivo)
    print("Contenido Pickle:\n")
    print(objeto)

# Escribir Pickle
nuevo_objeto = {
    "curso": "Python",
    "nivel": "Básico",
    "estudiantes": 30
}

with open("/content/drive/MyDrive/Intro Datos/nuevo_objeto.pkl", "wb") as archivo:
    pickle.dump(nuevo_objeto, archivo)
```

### Pregunta

¿Qué tipo de dato se podría almacenar con un Pickle?

**Respuesta:**
Pickle permite almacenar objetos de Python como listas, diccionarios, tuplas, DataFrames e incluso modelos de machine learning.
